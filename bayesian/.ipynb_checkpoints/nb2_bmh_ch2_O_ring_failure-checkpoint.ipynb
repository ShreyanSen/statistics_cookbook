# Challenge

Use bayesian modeling to determine the probability of failure as a function of temperature! We'll use the logistic function to generate a probability of failure as a function of temperature (plus a constant):


$P_{failure} = \frac{1}{1 + e^{\beta T}} + \alpha$

We change our alpha slightly from the one in their notebook! Our version is more interpretable so I like it more :) 

Then our defects are drawn from:

$ Defect \sim Bin(1, P_{failure})$

```python
# develop in a class to make it easier for reuse

class oRingAnalysis:
    
    def __init__(self, data, alpha_prior=70, beta_prior=0):
        self.data_array = data
        self.alpha_prior = alpha_prior
        self.beta_prior = beta_prior
        #self.run_model()

    def run_all(self):
        self.run_model()
        self.generate_posterior_fits()

    def plot_all(self):
        self.plot_param_hists()
        self.plot_probability_of_failure()
        
        

    def run_model(self):
        self.temperature_vals = self.data_array[:,0]
        self.defect_vals = self.data_array[:,1]
        with pm.Model() as model:
            beta = pm.Normal("beta", mu=self.beta_prior, tau=0.01)
            alpha = pm.Normal("alpha", mu=self.alpha_prior, tau=5) 
            p = pm.Deterministic("p", 1.0/(1.0 + at.exp(-beta*(self.temperature_vals - alpha))))
        
            observed = pm.Bernoulli("bernoulli_obs", p, observed=self.defect_vals)
        
            start = pm.find_MAP()
            step = pm.Metropolis()
            trace = pm.sample(120000, step=step, initvals=start)
    
        
        self.alpha_samples = np.concatenate(trace.posterior.alpha.data[:,100000::2])[:, None]  # best to make them 1d
        self.beta_samples = np.concatenate(trace.posterior.beta.data[:,100000::2])[:, None]

    def plot_param_hists(self):
        plt.figure(figsize=(5,5))
        plt.hist(self.alpha_samples)
        plt.title('Alpha Samples')

        plt.figure(figsize=(5,5))
        plt.hist(self.beta_samples)
        plt.title('Beta Samples')

    def generate_posterior_fits(self):
        """
        Generates samples for probability of failure as a function of time, by drawing samples from our joint posterior for alpha and beta 
        and then generating a set of probabilities from it via our logistic function, just like in the model
        """
        self.temperature_vals = self.data_array[:,0]
        self.temps = np.linspace(temperature_vals.min() - 5, temperature_vals.max()+5, 50)[:, None] # our horizontal axis

        def gen_p_t(alpha, beta):
            """
            probability of failure as a function of temperature given an alpha and beta
            """
            p_t = 1.0/(1.0+np.exp(-beta*(self.temps-alpha)))
            return p_t
        posterior_p_t_list = []
        for i in range(0, len(self.alpha_samples)):
            sampled_p_t = gen_p_t(self.alpha_samples[i], self.beta_samples[i])
            posterior_p_t_list.append(sampled_p_t)


        self.posterior_p_t_array = np.concatenate(posterior_p_t_list, axis=1)

    def plot_probability_of_failure(self):
        """
        The big reveal! We show our real data plus our mean and quantiles from the posterior bernoulli probabilities we generated
        """

        plt.figure(figsize=(20,8))
        plt.title('Failure Data Plus Modeled Failure Probability as Function of Temperature')
        plt.plot(self.temps, self.posterior_p_t_array.mean(axis=1), label='posterior mean failure probability')
        plt.plot(self.temps, np.quantile(self.posterior_p_t_array, 0.9, axis=1), linestyle='--', color='red', label='80th perc')
        plt.plot(self.temps, np.quantile(self.posterior_p_t_array, 0.1, axis=1), linestyle='--', color='red')
        plt.plot(self.temps, np.quantile(self.posterior_p_t_array, 0.99, axis=1), linestyle='--', color='pink', label='98th perc')
        plt.plot(self.temps, np.quantile(self.posterior_p_t_array, 0.01, axis=1), linestyle='--', color='pink')
        plt.xlabel('Temperature')
        plt.ylabel('Failure')
        
        plt.scatter(self.temperature_vals, self.defect_vals, color="k", s=50, alpha=0.5, label='failure data')
        plt.legend()

            
            

                   
```

```python
oring_obj = oRingAnalysis(data_array)
```

```python
oring_obj.run_all()
```

```python
oring_obj.plot_all()
```

### Miscalibrated Priors

What happens if we set our prior to 80 instead of 70? The results are so, so ugly! The model doesn't have the flexibility to move alpha over to where it's supposed to be. Instead it totally butchers beta. 

```python
oring_obj = oRingAnalysis(data_array, alpha_prior=80)
oring_obj.run_all()
oring_obj.plot_all()
```

### Fixing Our Priors

Thinking about our model a bit more, we do have reasonable beliefs about what alpha should be. It should be in the range where both failures and non-failures occur, as alpha corresponds mathematically to the temperature value where our sigmoid curve hits its 50% probability middle. But, what if we take a step back and go less informative, letting our model determine alpha for itself? I revamp model, changing alpha's prior to a uniform value with support over the entire data range--saying basically I have no idea where I expect it to be--and our alpha posterior converges to closer to 65 than 70! Furthermore, our sigmoid fit looks a lot better. Good stuff!

```python
# develop in a class to make it easier for reuse

class oRingAnalysisUnifPrior:
    
    def __init__(self, data, alpha_prior=70, beta_prior=0):
        self.data_array = data
        self.alpha_prior = alpha_prior
        self.beta_prior = beta_prior
        #self.run_model()

    def run_all(self):
        self.run_model()
        self.generate_posterior_fits()

    def plot_all(self):
        self.plot_param_hists()
        self.plot_probability_of_failure()
        
        

    def run_model(self):
        self.temperature_vals = self.data_array[:,0]
        self.defect_vals = self.data_array[:,1]
        with pm.Model() as model:
            beta = pm.Normal("beta", mu=self.beta_prior, tau=0.01)
            alpha = pm.Uniform("alpha", lower=50, upper=90)
            #alpha = pm.Normal("alpha", mu=self.alpha_prior, tau=5) 
            p = pm.Deterministic("p", 1.0/(1.0 + at.exp(-beta*(self.temperature_vals - alpha))))
        
            observed = pm.Bernoulli("bernoulli_obs", p, observed=self.defect_vals)
        
            start = pm.find_MAP()
            step = pm.Metropolis()
            trace = pm.sample(120000, step=step, initvals=start)
    
        
        self.alpha_samples = np.concatenate(trace.posterior.alpha.data[:,100000::2])[:, None]  # best to make them 1d
        self.beta_samples = np.concatenate(trace.posterior.beta.data[:,100000::2])[:, None]

    def plot_param_hists(self):
        plt.figure(figsize=(5,5))
        plt.hist(self.alpha_samples)
        plt.title('Alpha Samples')

        plt.figure(figsize=(5,5))
        plt.hist(self.beta_samples)
        plt.title('Beta Samples')

    def generate_posterior_fits(self):
        """
        Generates samples for probability of failure as a function of time, by drawing samples from our joint posterior for alpha and beta 
        and then generating a set of probabilities from it via our logistic function, just like in the model
        """
        self.temperature_vals = self.data_array[:,0]
        self.temps = np.linspace(temperature_vals.min() - 5, temperature_vals.max()+5, 50)[:, None] # our horizontal axis

        def gen_p_t(alpha, beta):
            """
            probability of failure as a function of temperature given an alpha and beta
            """
            p_t = 1.0/(1.0+np.exp(-beta*(self.temps-alpha)))
            return p_t
        posterior_p_t_list = []
        for i in range(0, len(self.alpha_samples)):
            sampled_p_t = gen_p_t(self.alpha_samples[i], self.beta_samples[i])
            posterior_p_t_list.append(sampled_p_t)


        self.posterior_p_t_array = np.concatenate(posterior_p_t_list, axis=1)

    def plot_probability_of_failure(self):
        """
        The big reveal! We show our real data plus our mean and quantiles from the posterior bernoulli probabilities we generated
        """

        plt.figure(figsize=(20,8))
        plt.title('Failure Data Plus Modeled Failure Probability as Function of Temperature')
        plt.plot(self.temps, self.posterior_p_t_array.mean(axis=1), label='posterior mean failure probability')
        plt.plot(self.temps, np.quantile(self.posterior_p_t_array, 0.9, axis=1), linestyle='--', color='red', label='80th perc')
        plt.plot(self.temps, np.quantile(self.posterior_p_t_array, 0.1, axis=1), linestyle='--', color='red')
        plt.plot(self.temps, np.quantile(self.posterior_p_t_array, 0.99, axis=1), linestyle='--', color='pink', label='98th perc')
        plt.plot(self.temps, np.quantile(self.posterior_p_t_array, 0.01, axis=1), linestyle='--', color='pink')
        plt.xlabel('Temperature')
        plt.ylabel('Failure')
        
        plt.scatter(self.temperature_vals, self.defect_vals, color="k", s=50, alpha=0.5, label='failure data')
        plt.legend()

            
            

                   
```

```python
oring_obj = oRingAnalysisUnifPrior(data_array)
oring_obj.run_all()
oring_obj.plot_all()
```
