# Multilevel Tadpoles

In this one we'll play with the multilevel tadpoles example used in Ch. 12 of Statistical Rethinking by Richard McElreath. Let's check out the difference between the weakly regularizing prior vs. the varying intercepts model. 

Background on the experiment:


A simple way to think about this model is to ask is: do we think the success probabilities for these tanks really come from the same distribution, or do they vary? 



```python
import pymc as pm
import pytensor.tensor as at
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d = pd.read_csv('Data/reedfrogs.csv', sep=",")
```

### Model A

Let $s_i, n_i, p_i$ represent number of frogs which survived, number of frogs originally present, and survival probability of frogs in the ith tank. Then we represent our first model as:

$ s_i \sim Bin(n_i, p_i) $


$ logit(p_i) \sim \alpha_i  $

$ \alpha_i \sim N(0,5) $



Let's just work out what the logit(p) = alpha implies, for greater clarity!

$ \ln(\frac{p_i}{1-p_i})\sim \alpha_i  $


$ \frac{p_i}{1-p_i} = e^{\alpha_i}$

$ p_i = (1-p_i) e^{\alpha_i}$

$ p_i (1 + e^{\alpha_i}) = e^{\alpha_i}$

$ p_i  = \frac{e^{\alpha_i}}{(1 + e^{\alpha_i})}$

$ p_i  = \frac{1}{(1 + e^{-\alpha_i})}$

Yup! The logit is the inverse logistic function just as we'd expect. Note that to make this more complicated and involve predictors, we set up alpha to also be a function of those predictors--which is what we did in the bayesian methods for hackers chapter 2 notebook in this directory. 






```python
class FrogsAnalysis():

    def __init__(self, data):
        RANDOM_SEED = 58
        self.rng = np.random.default_rng(RANDOM_SEED)
        self.data = data

    def run_model_a1(self):

        self.tank_n = self.data.density
        self.tank_survival = self.data.surv
        
        with pm.Model() as model:
            alpha = pm.Normal("alpha", mu = 0, tau = 5) 
            #alpha = pm.Normal("alpha", mu = 0, tau = 5, shape=self.data.shape[0]) # shape param allowed me to get a posterior alpha_i per tank!
            p = pm.Deterministic("p", 1.0/(1.0 + at.exp(-alpha)))
            
            observed = pm.Binomial("survived", n=self.tank_n, p=p, observed=self.tank_survival)

            step = pm.Metropolis()
            trace = pm.sample(10000, tune=5000, step=step, return_inferencedata=True)
            pm.sample_posterior_predictive(trace,  extend_inferencedata=True)

            # alternative formulation:
            #post_pred = pm.sample_posterior_predictive(trace, samples=1000, model=model)
            #self.model_a_post_pred = pm.sample_posterior_predictive(trace)
            
            # more on posterior predictive checks: https://www.pymc.io/projects/docs/en/stable/learn/core_notebooks/posterior_predictive.html

            self.model_a1_samples = trace
            #self.model_a_post_pred = post_pred

    def run_model_a2(self):

        self.tank_n = self.data.density
        self.tank_survival = self.data.surv
        
        with pm.Model() as model:
            #alpha = pm.Normal("alpha", mu = 0, tau = 5) 
            alpha = pm.Normal("alpha", mu = 0, tau = 5, shape=self.data.shape[0]) # shape param allowed me to get a posterior alpha_i per tank!
            p = pm.Deterministic("p", 1.0/(1.0 + at.exp(-alpha)))
            
            observed = pm.Binomial("survived", n=self.tank_n, p=p, observed=self.tank_survival)

            step = pm.Metropolis()
            trace = pm.sample(10000, tune=5000, step=step, return_inferencedata=True)
            pm.sample_posterior_predictive(trace,  extend_inferencedata=True)

            # alternative formulation:
            #post_pred = pm.sample_posterior_predictive(trace, samples=1000, model=model)
            #self.model_a_post_pred = pm.sample_posterior_predictive(trace)
            
            # more on posterior predictive checks: https://www.pymc.io/projects/docs/en/stable/learn/core_notebooks/posterior_predictive.html

            self.model_a2_samples = trace
            #self.model_a_post_pred = post_pred

    def run_model_b(self):

        self.tank_n = self.data.density
        self.tank_survival = self.data.surv
        
        with pm.Model() as model:
            #alpha = pm.Normal("alpha", mu = 0, tau = 5) 
            a = pm.Normal("a", mu=0, tau=3)
            sig = pm.HalfCauchy("sigma", 1)
            alpha = pm.Normal("alpha", mu = a, tau = sig, shape=self.data.shape[0]) # shape param allowed me to get a posterior alpha_i per tank!
            p = pm.Deterministic("p", 1.0/(1.0 + at.exp(-alpha)))
            
            observed = pm.Binomial("survived", n=self.tank_n, p=p, observed=self.tank_survival)

            step = pm.Metropolis()
            trace = pm.sample(10000, tune=5000, step=step, return_inferencedata=True)
            pm.sample_posterior_predictive(trace,  extend_inferencedata=True)

            # alternative formulation:
            #post_pred = pm.sample_posterior_predictive(trace, samples=1000, model=model)
            #self.model_a_post_pred = pm.sample_posterior_predictive(trace)
            
            # more on posterior predictive checks: https://www.pymc.io/projects/docs/en/stable/learn/core_notebooks/posterior_predictive.html

            self.model_b_samples = trace
            #self.model_a_post_pred = post_pred
    

    def plot_model_a2(self):
        plt.figure(figsize=(5,5))
        plt.hist(self.model_a2_samples.posterior.alpha.to_numpy().flatten())
        plt.title('Alpha Samples')

        plt.figure(figsize=(5,5))
        plt.hist(self.model_a2_samples.posterior.p.to_numpy().flatten())
        plt.title('P Samples')

    def plot_model_b(self):
        plt.figure(figsize=(5,5))
        plt.hist(self.model_b_samples.posterior.alpha.to_numpy().flatten())
        plt.title('Alpha Samples')

        plt.figure(figsize=(5,5))
        plt.hist(self.model_b_samples.posterior.p.to_numpy().flatten())
        plt.title('P Samples')
    
                
        
```

```python
stats_obj = FrogsAnalysis(d)
stats_obj.run_model_a1()
stats_obj.run_model_a2()
```

# Learning About What Was Observed

Originally, I had my alpha variable defined under my model definition without the shape parameter (have this as model A1). This yielded posteriors for alpha and p, but couldn't answer a question I wanted to answer, which was: conditional on what was observed in each tank, what is the posterior probability of survival in that tank? In other words, we know that we draw a p_i for each tank. But what does that p_i distribution look like at the end of the day?

I went down a deep rabbithole of posterior predictive checks, but these checks seemed suited to simulating data. They would draw a set of params and then sample from them to draw a simulated observation. But this didn't seem to answer my question. I didn't want to simulate new observations--I wanted the posterior survival probabilities for the original observations I passed in as data.

Changing the shape on alpha seemed to do the trick! This turned alpha into a multivariate normal. I don't know exactly how it works on the back end, but now the posterior has an additional dimension: 48, for the 48 original tanks. And for each tank I'm able to see a separate 4x10000 set of parameter samples, and when I average over them (for each tank) the estimated survival probabilities seem to match well against the actual drawn outcomes! On the other hand, when I just look at the posterior predictive values and get the average outcomes from them, they seem to be averages across all possible alpha draws based on N the number of frogs in each tank (aka they seem to not be conditioned on the actual outcome in each tank). 

```python
d['posterior_pred_p_model_a2'] =np.mean(stats_obj.model_a2_samples.posterior.p, axis=(0,1))
```

```python
d['posterior_pred_outcome_m_a2'] = np.mean(stats_obj.model_a2_samples.posterior_predictive.survived, axis=(0,1))
```

```python
d['posterior_pred_outcome_m_a1'] = np.mean(stats_obj.model_a1_samples.posterior_predictive.survived, axis=(0,1))
```

### Model B

Now let's take a look at what happens when we allow for adaptive regularization, letting alpha be drawn from a mixture of distributions rather than one distribution. We see that the samples for P look totally different under model B and model A2!

```python
stats_obj.run_model_b()
```

```python
d[['density','surv','propsurv','posterior_pred_p_model_a2', 'posterior_pred_p_model_b']] 
```

#### Comparison

Look at the difference! When we just set a N(0,5) prior on alpha, we over-regularized. By allowing an additional level of flexibility, and adaptive pooling of information, we see that the posterior probabilities are a lot closer to their empirical survival rates, meaning our model selected for less pooling of information (and a greater diversity in drawing alpha). This is reflected in the histograms, where under model B we see that the distribution for P has a lot more going on, with distinct regimes, one with a wider distribution centered around P = 0.4, another peaked around P = 0.9
