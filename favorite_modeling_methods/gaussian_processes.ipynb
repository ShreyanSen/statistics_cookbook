### generate data



```python

np.random.seed(42)
x = np.random.uniform(0, 10, size=15) # select 15 locations between 0 and 10

# A one dimensional column vector of inputs.
X = x[:, None]

y = np.random.normal(np.sin(x), 0.1) # generating our data: sin(x) plus random noise
plt.plot(x, y, 'o')
true_x = np.linspace(0, 10, 200)
plt.plot(true_x, np.sin(true_x), 'k--') # our real function: sin(x). nice and simple
plt.xlabel('x')
plt.ylabel('f(x)', rotation=0)
```

```python
X_grid = np.linspace(np.floor(x.min()), np.ceil(x.max()), 100)[:,None]
```

### GP with scikit learn

we start with as simple test case. we'll use scikit learn's fast implementation with a rbf kernel (same as squared exponential)
roughly following here:
https://scikit-learn.org/stable/auto_examples/gaussian_process/plot_gpr_noisy_targets.html#sphx-glr-auto-examples-gaussian-process-plot-gpr-noisy-targets-py

we'll include noise. first, we mispecify the noise resulting in a really poor fit. You can guess something's off because the length scale stepped to its upper bound...

```python

noise_std = 1
alpha=noise_std**2

kernel = 1 * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2))
gaussian_process = GaussianProcessRegressor(kernel=kernel, alpha=alpha, n_restarts_optimizer=9)
gaussian_process.fit(X, y)
gaussian_process.kernel_


mean_prediction, std_prediction = gaussian_process.predict(X_grid, return_std=True)

plt.plot(x, y, 'o')
true_x = np.linspace(0, 10, 200)
plt.plot(true_x, np.sin(true_x), 'k--') # our real function: sin(x). nice and simple
plt.xlabel('x')
plt.ylabel('f(x)', rotation=0)

plt.plot(X_grid, mean_prediction)
```

```python
gaussian_process.alpha # noise just way too high
```

#### improve initial params...

```python

noise_std = 0.1
alpha=noise_std**2

kernel = 1 * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2))
gaussian_process = GaussianProcessRegressor(kernel=kernel, alpha=alpha, n_restarts_optimizer=9)
gaussian_process.fit(X, y)
gaussian_process.kernel_


mean_prediction, std_prediction = gaussian_process.predict(X_grid, return_std=True)

plt.plot(x, y, 'o', label='observed data')
true_x = np.linspace(0, 10, 200)
plt.plot(true_x, np.sin(true_x), 'k--', label='true function') # our real function: sin(x). nice and simple
plt.xlabel('x')
plt.ylabel('f(x)', rotation=0)

plt.plot(X_grid, mean_prediction, label='gp mean')
plt.legend()
```

```python
gaussian_process.alpha # noise popped way low... makes sense given the data generating process
```

```python
gaussian_process.kernel # our original params
```

```python
gaussian_process.kernel_ # the fit params
```

#### plot draws from posterior

earlier we plotted the GP mean, but now let's draw sets of curves by sampling the GP posterior and plotting them

```python
num_draws = 100
samples = gaussian_process.sample_y(X_grid, num_draws)

for i in range(num_draws):
    plt.plot(X_grid, samples[:, i], color='green', alpha=0.1)

plt.title('Draws from GP Posterior')
plt.xlabel('x')
plt.ylabel('f(x)')
```

### GPs in PyMC

We can also do GPs in pymc. it does take a lot longer though, but the interoperability with other pymc based models is pretty great. note that how you set your priors has a lot to do with how long it takes...


https://www.pymc.io/projects/examples/en/latest/gaussian_processes/GP-Marginal.html
https://github.com/aloctavodia/BAP/blob/master/code/Chp7/07_Gaussian%20process.ipynb

let's use these two resources to set up and solve a simple example

first we'll start with pymc's marginal likelihood implementation, which assumes the data are created by an underlying plus gaussian noise. The GP requires a mean function and covariance function. We'll start with the most common use case: a squared exponential covariance. 



```python
with pm.Model() as model:
    # gamma hyperprior for lengthscale variable
    ell = pm.Gamma("ell", alpha=2, beta=0.5)
    # exponential squared covariance function
    cov = pm.gp.cov.ExpQuad(1, ls=ell)
    # GP prior
    gp = pm.gp.Marginal(cov_func=cov)

    # prior for gaussian noise
    sigma = pm.HalfNormal("sigma", 25)
    y_pred = gp.marginal_likelihood("y", X=X, y=y, sigma=sigma)
    
    trace = pm.sample(draws=1000, chains=2)
```

```python
az.plot_trace(trace)
```

```python
X_grid = np.linspace(np.floor(x.min()), np.ceil(x.max()), 100)[:,None]


with model:
    f_pred = gp.conditional('f_pred', X_grid)
```

```python
with model:
    pred_samples = pm.sample_posterior_predictive(trace, var_names=["f_pred"])
```

```python
pred_samples.posterior_predictive.f_pred[0].values[0,:]
```

```python
# Plot draws from posterior

for i in range(100):
    plt.plot(X_grid, pred_samples.posterior_predictive.f_pred[0].values[i,:], color='green', alpha=0.1)
```

## GP for Time Series Analysis

We'll follow these two resources for our tutorial:

https://www.pymc.io/projects/docs/en/v5.16.2/learn/core_notebooks/Gaussian_Processes.html

https://juanitorduz.github.io/gp_ts_pymc3/

Now, we generate a more complicated time series dataset. 



```python
# Generate seasonal variables. 

class ts_data:

    def __init__(self):
        self.gen_time_series_data()
        return
        
    def seasonal(self, t, amplitude, period):
        """Generate a sinusoidal curve."""
        return amplitude * np.sin((2*np.pi*t)/period) 
    
    def gen_time_series_data(self, data_length=300):
        self.data_length = data_length
        self.t = np.arange(0,data_length)
        self.seas_short = self.seasonal(self.t, 5, 4)
        self.seas_long =  self.seasonal(self.t, 10, 16)
        self.trend = 0.01*self.t
        self.random = np.random.normal(5, 10, data_length)
        self.y = self.seas_short + self.seas_long + self.trend + self.random

    def split_test_train_sequentially(self):

        split_val = 250
        self.test_size = split_val
        self.t_train = self.t[:split_val]
        self.t_test = self.t[split_val:]
        self.y_train = self.y[:split_val]
        self.y_test = self.y[split_val:]
        
        

    def plot_test_train(self):
        plt.scatter(self.t_train, self.y_train, label='train')
        plt.scatter(self.t_test, self.y_test, label='test')
        plt.legend()
```

```python
time_series = ts_data()
time_series.split_test_train_sequentially()
```

```python
time_series.plot_test_train()
```

```python
# let's model this! composing a three part GP

with pm.Model() as ts_model:
    # GP for first seasonal component    
    # gamma hyperprior for lengthscale variable
    ell_1 = pm.Gamma("ell_1", alpha=2, beta=0.5)
    # gamma hyperprior for period variable
    period_1 = pm.Gamma(name='period_1', alpha=10, beta=2) # mean is at 5, alpha/beta
    # periodic covariance function
    cov_1 = pm.gp.cov.Periodic(input_dim=1, period=period_1, ls=ell_1)
    # GP prior
    gp_1 = pm.gp.Marginal(cov_func=cov_1)

    # GP for second seasonal component
    ell_2 = pm.Gamma("ell_2", alpha=2, beta=0.5)
    period_2= pm.Gamma(name='period_2', alpha=40, beta=2) # mean is at 20
    cov_2 = pm.gp.cov.Periodic(input_dim=1, period=period_2, ls=ell_2)
    gp_2 = pm.gp.Marginal(cov_func=cov_2)

    # GP for a third non-seasonal component

    ell_3 = pm.Gamma("ell_3", alpha=2, beta=0.5)
    # gamma hyperprior for period variable
    cov_3 = pm.gp.cov.ExpQuad(1, ls=ell_3)
    gp_3 = pm.gp.Marginal(cov_func=cov_3)

    
    gp = gp_1+gp_2+gp_3
    # prior for gaussian noise
    sigma = pm.HalfNormal("sigma", 25)

    X_in = time_series.t_train.reshape(time_series.test_size, 1) # need to go from shape (k,) to (k,1)
    y_pred = gp.marginal_likelihood("y", X=X_in, y=time_series.y_train, sigma=sigma)
    
    trace = pm.sample(draws=1000, chains=2)

    
```

## pymc too slow; try sklearn

It was taking WAYYY too long to run on pymc. I don't really know why but sklearn is way, way faster for this. Let's see if we can solve this same problem with SKlearn. 

#### sensitivity to initial params

below are two sets of almost the same code, with the only differences being changes to the random seed, the noise_std term, and the length_scale for the first kernel.

What I noticed is that at too low of a noise, and too high of a length scale, we really don't get good convergence. The periodicity params for the two seasonal kernels don't converge at all to reality (4 and 16). However, once we load it with enough noise (2) to be flexible and reduce the length scales, I see that I get reliable convergence to the right periodicities.

#### results

The results in the plot don't look too bad! However, we miss quite a bit--like all the big peaks. If we were, for instance, trading off those peaks, then this would be a very poor model indeed. I like that it correctly identified the periodicities, but I don't like that I had to fiddle a bit with initial params to converge to that right answer. Probably if I understood the ExpSineSquared's length scale param a bit better I might know a priori how to set it--as of now, I don't understand it fully (since the intuition changes--there is already "action at a distance" happening from the periodicity...)

```python
np.random.seed(40)

noise_std = .1
alpha=noise_std**2


# longer seasonal kernel
kernel_1 = ExpSineSquared(length_scale = 20, periodicity=20, length_scale_bounds=(1e-2, 1e2), periodicity_bounds=(1e-2, 1e2))
# shorter seasonal kernel
kernel_2 = ExpSineSquared(length_scale = 5, periodicity=5, length_scale_bounds=(1e-2, 1e2), periodicity_bounds=(1e-2, 1e2))
kernel_3 = RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2))
kernel = kernel_1 + kernel_2 + kernel_3
gaussian_process = GaussianProcessRegressor(kernel=kernel, alpha=alpha, n_restarts_optimizer=9)

X_in = time_series.t_train.reshape(time_series.test_size, 1) # need to go from shape (k,) to (k,1)


gaussian_process.fit(X_in, time_series.y_train)
gaussian_process.kernel_
```
