```python
import os

import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc as pm
import pytensor.tensor as pt
```

```python
import libpysal
from geopandas import read_file
az.style.use("arviz-darkgrid")
```

```python
data_url = 'https://raw.githubusercontent.com/pymc-devs/pymc-examples/refs/heads/main/examples/data/scotland_lips_cancer.csv'
slc_df = pd.read_csv(data_url)
```

```python
slc_df.head()
```

CANCER here refers to counts of the number of lip cancer cases per county, and the model used is additive against CEXP which is the expected number of lip cancer cases per county given the populations and ages of the county. AFF refers to percent population engaged in agriculture, forestry, or fishing. 

The model looks like:

$B_0, B_1 \sim N(0, a)$

$\phi_i | \phi_{j\sim i} \sim N(\alpha \sum_{j=1}^{n_i}w_{ij}\phi_j, \sigma_i^2)$

$\log \lambda_i = B_0 + B_1 x_i + \phi_i + \log E_i$

$y_i \sim Poisson(\lambda_i) $

Here $E_i$ and $x_i$ represent predictors (CEXP and AFF respectively) for each county, and we have county independent $B$ factors for slope against predictors plus intercept. Finally, we have our spatially autoregressive term $\phi_i$ which draws with a county dependent variance term and a mean that's given by a weighted sum of the $\phi$ values at all neighboring counties. We sum those up to get our rate parameter $\lambda$ per county (with some logarithms ofc) and model our actual cancer counts with this poisson process.


In our model the weights are going to be 1 if j is a neighboring county and 0 if not. So if we know all the neighbor's parameters, we expect ours to be the sum of those. Visualizing that is kind of interesting. If your neighbors are all positive you're deeply positive. If half your neighbors are negative and half are positive with equal magnitude, you're around 0--neutral, transition county. 

```python
# data formatting
# copying this straight from the tutorial because low yield effort :)

# observed cancer counts
y = slc_df["CANCER"].values

# number of observations
N = len(y)

# expected cancer counts E for each county: this is calculated using age-standardized rates of the local population
E = slc_df["CEXP"].values
logE = np.log(E)

x = slc_df["AFF"].values / 10.0


# adjacency matrix

# spatial adjacency information: column `ADJ` contains list entries which are preprocessed to obtain adj as a list of lists
adj = (
    slc_df["ADJ"].apply(lambda x: [int(val) for val in x.strip("][").split(",")]).to_list()
)

# change to Python indexing (i.e. -1)
for i in range(len(adj)):
    for j in range(len(adj[i])):
        adj[i][j] = adj[i][j] - 1

# storing the adjacency matrix as a two-dimensional np.array
adj_matrix = np.zeros((N, N), dtype="int32")

for area in range(N):
    adj_matrix[area, adj[area]] = 1
```

### First Model: No Spatial Autocorrelation

In this model we replace $\phi_i | \phi_j$ with parameter $\theta$ which is independent for each spatial unit 

The tutorial picks really specific priors, including tiny tiny priors on beta0 and beta1, possibly because we're living in log space here. 
I will relax those assumptions and see what we get.

```python
with pm.Model(coords={"area_idx": np.arange(N)}) as independent_model:
    beta0 = pm.Normal("beta0", mu=0, tau=1)
    beta1 = pm.Normal("beta1", mu=0, tau=1)
    # variance parameter of the independent random effect
    tau_ind = pm.Gamma("tau_ind", alpha=5, beta=1)

    # independent random effect
    theta = pm.Normal("theta", mu=0, tau=tau_ind, dims="area_idx")

    # exponential of the linear predictor -> the mean of the likelihood
    mu = pm.Deterministic("mu", pt.exp(logE + beta0 + beta1 * x + theta), dims="area_idx")

    # likelihood of the observed data
    y_i = pm.Poisson("y_i", mu=mu, observed=y, dims="area_idx")

    # saving the residual between the observation and the mean response for the area
    res = pm.Deterministic("res", y - mu, dims="area_idx")

    # sampling the model
    independent_idata = pm.sample(2000, tune=2000)
```

Let's look at the mean exponentiated theta value, which represents a multiplicative scaling factor for each county
We see that they're assigned much like the original cancer counts were.

```python
spat_df["exp_theta"] = np.exp(independent_idata["posterior"]["theta"].mean(dim=["chain", "draw"]))
independent_map = spat_df.plot(
    column="exp_theta",
    scheme="QUANTILES",
    k=5,
    cmap="BuPu",
    legend=True,
    legend_kwds={"loc": "center left", "bbox_to_anchor": (1, 0.5)},
)
```

We also see a pretty strong effect coming from our AFF variable

```python
independent_idata["posterior"]["beta1"].mean(dim=["chain", "draw"])
```

```python
np.quantile(independent_idata["posterior"]["beta1"], 0.95)
```

```python
np.quantile(independent_idata["posterior"]["beta1"], 0.05)
```

But what about the residuals?
We see they are not spatially random, but distributed the same way the original data were and with some spatial clustering. So let's try the spatial dependency. 

```python
spat_df["INDEPENDENT_RES"] = independent_idata["posterior"]["res"].mean(dim=["chain", "draw"])
independent_map = spat_df.plot(
    column="INDEPENDENT_RES",
    scheme="QUANTILES",
    k=5,
    cmap="BuPu",
    legend=True,
    legend_kwds={"loc": "center left", "bbox_to_anchor": (1, 0.5)},
)
```

### Second Model: Spatial Autocorrelation

So this is much easier to write because pymc already has a module in place for CAR. Otherwise, we'd have to handle the neighboring indices the same way we did for the AR process, which sounds like a bit of a nightmare. In this model, we'll fit an independent effect per county as well as a CAR process. All we need on the generative side is to pass it an adjacency matrix, and remember--in the pymc framework as long as we can generate an equivalent dataset from our set of priors, and we have all our likelihoods, then we can work backwards and get modified parameters. 

They give two examples, one with and one without a prior on their fixed weight param. Unfortunately, both go very wrong for us. Sampling takes forever and the models get stuck somewhere. Hard to say if it's an implementation problem or something else.

We replace the CAR component with ICAR, which is the same model more or less but setting alpha=1.
In this case, the model is able to converge, so let's take a look at the results.

```python
with pm.Model(coords={"area_idx": np.arange(N)}) as fixed_spatial_model:
    beta0 = pm.Normal("beta0", mu=0.0, tau=1.0e-5)
    beta1 = pm.Normal("beta1", mu=0.0, tau=1.0e-5)
    # variance parameter of the independent random effect
    tau_ind = pm.Gamma("tau_ind", alpha=3.2761, beta=1.81)
    # variance parameter of the spatially dependent random effects
    tau_spat = pm.Gamma("tau_spat", alpha=1.0, beta=1.0)

    # area-specific model parameters
    # independent random effect
    theta = pm.Normal("theta", mu=0, tau=tau_ind, dims="area_idx")
    # spatially dependent random effect, alpha fixed
    #phi = pm.CAR("phi", mu=np.zeros(N), tau=tau_spat, alpha=0.95, W=adj_matrix, dims="area_idx")
    
    phi = pm.ICAR("phi", W=adj_matrix, dims="area_idx")
    # exponential of the linear predictor -> the mean of the likelihood
    mu = pm.Deterministic("mu", pt.exp(logE + beta0 + beta1 * x + theta + phi), dims="area_idx")

    # likelihood of the observed data
    y_i = pm.Poisson("y_i", mu=mu, observed=y, dims="area_idx")

    # saving the residual between the observation and the mean response for the area
    res = pm.Deterministic("res", y - mu, dims="area_idx")

    # sampling the model
    fixed_spatial_idata = pm.sample(2000, tune=1000)
```

```python
fixed_spatial_idata.posterior.phi
```

```python
spat_df["exp_theta_icar"] = np.exp(fixed_spatial_idata["posterior"]["theta"].mean(dim=["chain", "draw"]))
independent_map = spat_df.plot(
    column="exp_theta_icar",
    scheme="QUANTILES",
    k=5,
    cmap="BuPu",
    legend=True,
    legend_kwds={"loc": "center left", "bbox_to_anchor": (1, 0.5)},
)
```

Right off the bat we see that our independent theta term actually looks spatially independent now. Let's see what our correlated params look like.

```python
spat_df["exp_phi_icar"] = np.exp(fixed_spatial_idata["posterior"]["phi"].mean(dim=["chain", "draw"]))
independent_map = spat_df.plot(
    column="exp_phi_icar",
    scheme="QUANTILES",
    k=5,
    cmap="BuPu",
    legend=True,
    legend_kwds={"loc": "center left", "bbox_to_anchor": (1, 0.5)},
)
```

And look at that. We've got our beautiful spatial autocorrelation terms. How are our residuals looking?

```python
spat_df["ICAR_RES"] = fixed_spatial_idata["posterior"]["res"].mean(dim=["chain", "draw"])
independent_map = spat_df.plot(
    column="ICAR_RES",
    scheme="QUANTILES",
    k=5,
    cmap="BuPu",
    legend=True,
    legend_kwds={"loc": "center left", "bbox_to_anchor": (1, 0.5)},
)
```

### Discussion

Well, we've definitely done a better job here. The spatial clumping of the residuals is broken up. However, our residuals still match our thetas fairly closely. That means that our residuals are not iid, there is still structure in them correlated to our independent effects. 

We should also take a closer look at the effects of the scaling factor. Our original tutorial kind of skimmed over this, but mentioned the same thing the second one does (https://www.pymc.io/projects/examples/en/latest/spatial/nyc_bym.html#scaling-factor)--which is that the priors are used to control an issue where, basically, non-standardized data can give illusive effects. But I have to admit, I find the details pretty confusing. Here is all the first tutorial has to say about it:

"Side note: Here we explain the prior’s used for the precision of the two random effect terms. As we have two random effects
and
for each
, they are independently unidentifiable, but the sum
is identifiable. However, by scaling the priors of this precision in this manner, one may be able to interpret the proportion of variance explained by each of the random effects."


Overall, it's nice to see that spatial autocorrelation terms can syntactically be included in pymc fairly easily. However, adequately modeling and managing the nuance here is of course a continuous journey, just as it is in the case of temporal autocorrelation.
