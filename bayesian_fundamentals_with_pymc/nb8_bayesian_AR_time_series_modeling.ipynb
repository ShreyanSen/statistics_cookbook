### Generate + Solve AR Process

We'll start with an autoregressive process

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pymc as pm
import arviz as az
import xarray as xr
#from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
az.style.use('arviz-darkgrid')
```

```python
np.random.seed(34)
def sim_ar(intercept = 10, thetas = [-0.9,0.9], noise_var = 0.3, steps=500):
    X = np.zeros(steps)
    order = len(thetas) # by default it's 2; aka AR(2) process
    X[:order] = intercept
    for i in range(order, steps):
        X[i] = np.sum(thetas*np.flip(X[(i-order):i])) + np.random.normal(0, noise_var)
    return X
    
```

```python
data = sim_ar(thetas = [-0.9,0.1])
```

## Bayesian AR Model

Now, let's do some Bayesian modeling. 

```python
# setting up dict of priors
priors = {
    "coefs": {"mu": [0, 0], "sigma": [1, 1], "size": 2},
    "sigma": 1,
    "init": {"mu": 10, "sigma": 3, "size": 2},
}
#conceit is we pretty much know the initial value;
# we need a prior each for each coefficient; here we'll go AR(2)

# split data into test and train
split = 300
data_train = data[0:split]
data_test = data[split:]
#t_data = list(range(split))
t_data = np.arange(split)
```

```python
with pm.Model() as AR:
    np.random.seed(82341)
    AR.add_coord("obs_id", t_data, mutable=True) # coords, mutable for prediction

    # data containers for time and 
    t = pm.MutableData("t", t_data, dims="obs_id")
    y = pm.MutableData("y", data_train, dims="obs_id")

    # now we build our priors for the AR model parameters

    # rho in the API, one per autoregression coefficient
    coefs = pm.Normal("coefs", priors["coefs"]["mu"], priors["coefs"]["sigma"])
    # standard deviation of our normal noise term
    sigma = pm.HalfNormal("sigma", priors["sigma"])
    # We need one init variable for each lag, hence size is variable too
    init = pm.Normal.dist(
        priors["init"]["mu"], priors["init"]["sigma"], size=priors["init"]["size"]
    )

    ar = pm.AR(
        "ar",
        coefs,
        sigma=sigma,
        init_dist=init,
        constant=False,
        dims="obs_id",
    )
    outcome = pm.Normal("likelihood", mu=ar, sigma=sigma, observed=y, dims="obs_id")

    idata_ar = pm.sample_prior_predictive()
    idata_ar.extend(pm.sample(2000, random_seed=100, target_accept=0.95))
    idata_ar.extend(pm.sample_posterior_predictive(idata_ar))
```

## Time Series Projections With PyMC

I want to demonstrate to myself that we can run prediction forward in time by using the same parameters (this is a syntax issue not a theory issue). I'll do it by running a very simple linear regression, fitting it over a training range, and then extending to a test range


```python
timesteps = 200
t = np.arange(timesteps)
theta = 5
noise = np.random.normal(0,20, timesteps)
y = theta*t + noise
```

```python
plt.plot(t, y, label="data")
plt.title("Data No Projections")
plt.legend()
```

```python
with pm.Model() as linmod:
    linmod.add_coords({"obs_id": range(timesteps)})
    
    time = pm.MutableData("t", t, dims="obs_id")
    vals = pm.MutableData("y", y, dims="obs_id")
    params = pm.Normal("params", 0,10)
    sigma = pm.HalfNormal("sigma", 5)
    mu = params*time
    obs = pm.Normal("obs", mu, sigma, observed=vals, dims="obs_id")


    #trace = pm.sample(draws=1000,tune=1000)
    
```

```python
projection_steps = 200
with linmod:
    linmod.add_coords({"obs_id_fut": range(timesteps,timesteps+projection_steps,1 )})

    t_fut = pm.MutableData(
        "t_fut", list(range(timesteps, timesteps+projection_steps, 1)), dims="obs_id_fut"
    )
    mu = pm.Deterministic("mu_fut", params * t_fut, dims="obs_id_fut")
    yhat_fut = pm.Normal("yhat_fut", mu=mu, sigma=sigma, dims="obs_id_fut")

    # use the updated values and predict outcomes and probabilities:
    trace_future =  pm.sample(draws=1000,tune=1000)
```

## Adding Trend and Seasonal Terms

To complicate our picture more let's add some trend and seasonal terms to the data and then see if we're able to recover them. For trend we'll just add a term that's linear with time. This follows roughly what the tutorial is doing but let's model it in our own way.

```python
class gen_synth_data:

    def __init__(self, t_train=300,t_test=200 ):
        self.t_train = t_train
        self.t_test = t_test
        self.t_length = self.t_train+self.t_test
        np.random.seed(31)
        self.t = np.arange(0,self.t_length)
        
        
            
    
    def build_custom_data(self):

        self.ar_comp = self.sim_ar(thetas = [-0.8,0.2])
        self.linear_comp = self.sim_lin(slope=0.01, intercept=0)
        self.seasonal_comp = self.sim_cyclic(amplitude = 4, frequency=(1/50), phase=0)
        self.noise = np.random.normal(0,1,self.t_length)
        self.data = self.ar_comp+self.linear_comp+self.seasonal_comp + self.noise

    def build_data_no_ar(self):

        self.linear_comp = self.sim_lin(slope=0.01, intercept=0)
        self.seasonal_comp = self.sim_cyclic(amplitude = 4, frequency=(1/50), phase=0)

        self.data = self.linear_comp+self.seasonal_comp

    def plot_custom_data(self):
        
        plt.figure(figsize=(20,5))
        plt.plot(self.t, self.data)
        plt.title('Data')
    
        plt.figure(figsize=(20,5))
        plt.plot(self.t, self.ar_comp)
        plt.title('AR')
    
        plt.figure(figsize=(20,5))
        plt.plot(self.t, self.linear_comp)
        plt.title('linear')
    
        plt.figure(figsize=(20,5))
        plt.plot(self.t, self.seasonal_comp)
        plt.title('seasonal')

    
    def sim_cyclic(self, amplitude, frequency, phase):
        cyc = amplitude*np.sin(2*np.pi*frequency*(self.t + phase))
        return cyc
        
    def sim_lin(self, slope, intercept):
        lin_val = slope*self.t + intercept
        return lin_val
        
    def sim_ar(self, intercept = 10, thetas = [-0.9,0.9], noise_var = 0.3):
        self.ar_order = len(thetas) # by default it's 2; aka AR(2) process
        X = np.zeros(self.t_length)
        X[:self.ar_order] = intercept
        for i in range(self.ar_order, self.t_length):
            X[i] = np.sum(thetas*np.flip(X[(i-self.ar_order):i])) + np.random.normal(0, noise_var)
        return X
```

```python
class sbts_model:

    def __init__(self, data_obj, priors = {
            "coefs": {"mu": [0, 0], "sigma": [1, 1], "size": 2},
            "sigma": 1,
            "init": {"mu": 10, "sigma": 3, "size": 2},
            "alpha":{"mu":0, "sigma":5},
            "beta":{"mu":0.2, "sigma":1},
            "amp":{"mu":5, "sigma":5},
            "freq":{"alpha":1,"beta":20}
        }):

        self.priors = priors

        self.data_obj = data_obj
        np.random.seed(9274)

        self.ar_order = len(priors["coefs"]["mu"]) #data_obj.ar_order
        self.data = data_obj.data
        
        self.train_length = data_obj.t_train
        self.test_length = data_obj.t_test
        self.total_length = data_obj.t_length
        
        self.t_train = np.arange(self.train_length)
        self.t_test = np.arange(self.train_length, self.total_length)

        self.data_train = self.data[0:self.train_length]
        self.data_test = self.data[self.train_length:self.total_length]
        
    def run_model(self):

        with pm.Model() as SBTS:
        
            SBTS.add_coord("obs_id", self.t_train, mutable=True) # coords, mutable for prediction
        
            # data containers for time and 
            t = pm.MutableData("t", self.t_train, dims="obs_id")
            y = pm.MutableData("y", self.data_train, dims="obs_id")
        
            # now we build our priors for the AR model parameters
        
            # rho in the API, one per autoregression coefficient
            coefs = pm.Normal("coefs", self.priors["coefs"]["mu"], self.priors["coefs"]["sigma"])
            # standard deviation of our normal noise term
            sigma = pm.HalfNormal("sigma", self.priors["sigma"])
            # We need one init variable for each lag, hence size is variable too
            init = pm.Normal.dist(
                self.priors["init"]["mu"], self.priors["init"]["sigma"], size=self.priors["init"]["size"]
            )
        
            ar = pm.AR(
                "ar",
                coefs,
                sigma=sigma,
                init_dist=init,
                constant=False,
                dims="obs_id",
            )

            # TREND params
            
            alpha = pm.Normal("alpha", self.priors["alpha"]["mu"], self.priors["alpha"]["sigma"])
            beta = pm.Normal("beta", self.priors["beta"]["mu"], self.priors["beta"]["sigma"])
            trend = pm.Deterministic("trend", alpha + beta * t, dims="obs_id")

            # SEASONAL params

            # beta prior for frequency since it's between 0 and 1; let's give ourselves an easier
            freq = pm.Beta("freq", alpha=self.priors["freq"]["alpha"], beta = self.priors["freq"]["beta"])
            # kind of a crappy choice of prior for amplitude but we could do worse
            amp = pm.Normal("amp", mu=self.priors["amp"]["mu"], sigma = self.priors["amp"]["sigma"])
            phase = pm.Uniform("phase", lower = -0.5, upper=0.5) # in this case we can hard code phase...we just pick uniform over the range
            
            seasonal = pm.Deterministic("seasonal", amp*np.sin(2*np.pi*freq*(t + phase)), dims="obs_id")
            
            mu = ar + trend + seasonal
            
            outcome = pm.Normal("likelihood", mu=mu, sigma=sigma, observed=y, dims="obs_id")
        
            self.idata = pm.sample_prior_predictive()
            self.idata.extend(pm.sample(3000, random_seed=100, target_accept=0.95))
            self.idata.extend(pm.sample_posterior_predictive(self.idata))

    def run_model_plus_forecast(self):

        with pm.Model() as SBTS:
        
            SBTS.add_coord("obs_id", self.t_train, mutable=True) # coords, mutable for prediction
        
            # data containers for time and 
            t = pm.MutableData("t", self.t_train, dims="obs_id")
            y = pm.MutableData("y", self.data_train, dims="obs_id")
        
            # now we build our priors for the AR model parameters
        
            # rho in the API, one per autoregression coefficient
            coefs = pm.Normal("coefs", self.priors["coefs"]["mu"], self.priors["coefs"]["sigma"])
            # standard deviation of our normal noise term
            sigma = pm.HalfNormal("sigma", self.priors["sigma"])
            # We need one init variable for each lag, hence size is variable too
            init = pm.Normal.dist(
                self.priors["init"]["mu"], self.priors["init"]["sigma"], size=self.priors["init"]["size"]
            )
        
            ar = pm.AR(
                "ar",
                coefs,
                sigma=sigma,
                init_dist=init,
                constant=False,
                dims="obs_id",
            )

            # TREND params
            
            alpha = pm.Normal("alpha", self.priors["alpha"]["mu"], self.priors["alpha"]["sigma"])
            beta = pm.Normal("beta", self.priors["beta"]["mu"], self.priors["beta"]["sigma"])
            trend = pm.Deterministic("trend", alpha + beta * t, dims="obs_id")

            # SEASONAL params

            # beta prior for frequency since it's between 0 and 1; let's give ourselves an easier
            freq = pm.Beta("freq", alpha=self.priors["freq"]["alpha"], beta = self.priors["freq"]["beta"])
            # kind of a crappy choice of prior for amplitude but we could do worse
            amp = pm.Normal("amp", mu=self.priors["amp"]["mu"], sigma = self.priors["amp"]["sigma"])
            phase = pm.Uniform("phase", lower = -0.5, upper=0.5) # in this case we can hard code phase...we just pick uniform over the range
            
            seasonal = pm.Deterministic("seasonal", amp*np.sin(2*np.pi*freq*(t + phase)), dims="obs_id")
            
            mu = ar + trend + seasonal
            
            outcome = pm.Normal("likelihood", mu=mu, sigma=sigma, observed=y, dims="obs_id")
        
            self.idata = pm.sample_prior_predictive()
            self.idata.extend(pm.sample(3000, random_seed=100, target_accept=0.95))
            self.idata.extend(pm.sample_posterior_predictive(self.idata))

            num_lags = self.ar_order
            # for now we let the code be inefficiently named and structured (due to copied from cell way above) instead focusing on no typos
            with SBTS:
                SBTS.add_coords({"obs_id_fut_lagged": range(len(self.data_train) - num_lags,len(self.data_test)+len(self.data_train),1 )})
                SBTS.add_coords({"obs_id_fut": self.t_test})

                t_fut = pm.MutableData(
                    "t_fut", self.t_test, dims="obs_id_fut"
                )
                ar_fut = pm.AR(
                    "ar_fut",
                    init_dist=pm.DiracDelta.dist(ar[..., -num_lags]),
                    rho=coefs,
                    sigma=sigma,
                    constant=False,
                    dims="obs_id_fut_lagged",
                )

                trend_fut = pm.Deterministic("trend_fut", alpha + beta * t_fut, dims="obs_id_fut")
                seasonal_fut = pm.Deterministic("seasonal_fut", amp*np.sin(2*np.pi*freq*(t_fut + phase)), dims="obs_id_fut")

                mu = ar_fut[num_lags:] + trend_fut + seasonal_fut
                yhat_fut = pm.Normal("yhat_fut", mu=mu, sigma=sigma, dims="obs_id_fut")
            
                # use the updated values and predict outcomes and probabilities:
                self.idata_preds = pm.sample_posterior_predictive(
                    self.idata, var_names=["likelihood", "yhat_fut"], predictions=True, random_seed=100
                )

    def run_no_ar_model_plus_forecast(self):

        with pm.Model() as SBTS:
        
            SBTS.add_coord("obs_id", self.t_train, mutable=True) # coords, mutable for prediction
        
            # data containers for time and 
            t = pm.MutableData("t", self.t_train, dims="obs_id")
            y = pm.MutableData("y", self.data_train, dims="obs_id")
        
            # TREND params
            
            alpha = pm.Normal("alpha", self.priors["alpha"]["mu"], self.priors["alpha"]["sigma"])
            beta = pm.Normal("beta", self.priors["beta"]["mu"], self.priors["beta"]["sigma"])
            trend = pm.Deterministic("trend", alpha + beta * t, dims="obs_id")

            # SEASONAL params

            # beta prior for frequency since it's between 0 and 1; let's give ourselves an easier
            freq = pm.Beta("freq", alpha=self.priors["freq"]["alpha"], beta = self.priors["freq"]["beta"])
            # kind of a crappy choice of prior for amplitude but we could do worse
            amp = pm.Normal("amp", mu=self.priors["amp"]["mu"], sigma = self.priors["amp"]["sigma"])
            phase = pm.Uniform("phase", lower = -0.5, upper=0.5) # in this case we can hard code phase...we just pick uniform over the range
            
            seasonal = pm.Deterministic("seasonal", amp*np.sin(2*np.pi*freq*(t + phase)), dims="obs_id")
            
            mu = trend + seasonal
            
            sigma = pm.HalfNormal("sigma", self.priors["sigma"])
            outcome = pm.Normal("likelihood", mu=mu, sigma=sigma, observed=y, dims="obs_id")
        
            self.idata = pm.sample_prior_predictive()
            self.idata.extend(pm.sample(2000, random_seed=100, target_accept=0.95))
            self.idata.extend(pm.sample_posterior_predictive(self.idata))

            num_lags = self.ar_order
            # for now we let the code be inefficiently named and structured (due to copied from cell way above) instead focusing on no typos
            with SBTS:
                SBTS.add_coords({"obs_id_fut": self.t_test})

                t_fut = pm.MutableData(
                    "t_fut", self.t_test, dims="obs_id_fut"
                )

                trend_fut = pm.Deterministic("trend_fut", alpha + beta * t_fut, dims="obs_id_fut")
                seasonal_fut = pm.Deterministic("seasonal_fut", amp*np.sin(2*np.pi*freq*(t_fut + phase)), dims="obs_id_fut")

                mu = trend_fut + seasonal_fut
                yhat_fut = pm.Normal("yhat_fut", mu=mu, sigma=sigma, dims="obs_id_fut")
            
                # use the updated values and predict outcomes and probabilities:
                self.idata_preds = pm.sample_posterior_predictive(
                    self.idata, var_names=["likelihood", "yhat_fut"], predictions=True, random_seed=100
                )


    def plot_model(self):
        plt.figure(figsize=(20,5))
        self.idata.posterior_predictive.likelihood.mean(["chain", "draw"]).plot(label="Posterior Mean AR level")
        plt.plot(self.data_train,'o',color="black", markersize=2, label="Observed Data")
        plt.legend()

        plt.figure(figsize=(20,5))
        plt.plot(self.idata.posterior.trend.mean(['chain','draw']), label='Trend')
        plt.legend()

    def plot_predictions(self):
        plt.figure(figsize=(20,4))
        plt.scatter(self.idata.constant_data.t, self.idata.constant_data.y, label='training data')
        plt.scatter(self.idata_preds.predictions_constant_data.t_fut, self.idata_preds.predictions.yhat_fut.mean(['chain','draw']), label='predicted values')
        plt.scatter(self.data_obj.t, self.data_obj.data, alpha=0.3, label='true values')
```

```python
class sbts_model_no_ar:

    def __init__(self, data_obj, priors = {
            "sigma": 1,
            "init": {"mu": 10, "sigma": 3, "size": 2},
            "alpha":{"mu":0, "sigma":5},
            "beta":{"mu":0.2, "sigma":1},
            "amp":{"mu":5, "sigma":5},
            "freq":{"alpha":1,"beta":20}
        }):

        self.priors = priors

        self.data_obj = data_obj
        np.random.seed(9274)

        self.data = data_obj.data
        
        self.train_length = data_obj.t_train
        self.test_length = data_obj.t_test
        self.total_length = data_obj.t_length
        
        self.t_train = np.arange(self.train_length)
        self.t_test = np.arange(self.train_length, self.total_length)

        self.data_train = self.data[0:self.train_length]
        self.data_test = self.data[self.train_length:self.total_length]
        

    def run_no_ar_model_plus_forecast(self):

        with pm.Model() as SBTS:
        
            SBTS.add_coord("obs_id", self.t_train, mutable=True) # coords, mutable for prediction
        
            # data containers for time and 
            t = pm.MutableData("t", self.t_train, dims="obs_id")
            y = pm.MutableData("y", self.data_train, dims="obs_id")
        
            # TREND params
            
            alpha = pm.Normal("alpha", self.priors["alpha"]["mu"], self.priors["alpha"]["sigma"])
            beta = pm.Normal("beta", self.priors["beta"]["mu"], self.priors["beta"]["sigma"])
            trend = pm.Deterministic("trend", alpha + beta * t, dims="obs_id")

            # SEASONAL params

            # beta prior for frequency since it's between 0 and 1; let's give ourselves an easier
            freq = pm.Beta("freq", alpha=self.priors["freq"]["alpha"], beta = self.priors["freq"]["beta"])
            # kind of a crappy choice of prior for amplitude but we could do worse
            amp = pm.Normal("amp", mu=self.priors["amp"]["mu"], sigma = self.priors["amp"]["sigma"])
            phase = pm.Uniform("phase", lower = -0.5, upper=0.5) # in this case we can hard code phase...we just pick uniform over the range
            
            seasonal = pm.Deterministic("seasonal", amp*np.sin(2*np.pi*freq*(t + phase)), dims="obs_id")
            
            mu = trend + seasonal
            
            sigma = pm.HalfNormal("sigma", self.priors["sigma"])
            outcome = pm.Normal("likelihood", mu=mu, sigma=sigma, observed=y, dims="obs_id")
        
            self.idata = pm.sample_prior_predictive()
            self.idata.extend(pm.sample(2000, random_seed=100, target_accept=0.95))
            self.idata.extend(pm.sample_posterior_predictive(self.idata))
            # for now we let the code be inefficiently named and structured (due to copied from cell way above) instead focusing on no typos
            with SBTS:
                SBTS.add_coords({"obs_id_fut": self.t_test})

                t_fut = pm.MutableData(
                    "t_fut", self.t_test, dims="obs_id_fut"
                )

                trend_fut = pm.Deterministic("trend_fut", alpha + beta * t_fut, dims="obs_id_fut")
                seasonal_fut = pm.Deterministic("seasonal_fut", amp*np.sin(2*np.pi*freq*(t_fut + phase)), dims="obs_id_fut")

                mu = trend_fut + seasonal_fut
                yhat_fut = pm.Normal("yhat_fut", mu=mu, sigma=sigma, dims="obs_id_fut")
            
                # use the updated values and predict outcomes and probabilities:
                self.idata_preds = pm.sample_posterior_predictive(
                    self.idata, var_names=["likelihood", "yhat_fut"], predictions=True, random_seed=100
                )


    def plot_model(self):
        plt.figure(figsize=(20,5))
        self.idata.posterior_predictive.likelihood.mean(["chain", "draw"]).plot(label="Posterior Mean AR level")
        plt.plot(self.data_train,'o',color="black", markersize=2, label="Observed Data")
        plt.legend()

        plt.figure(figsize=(20,5))
        plt.plot(self.idata.posterior.trend.mean(['chain','draw']), label='Trend')
        plt.legend()

    def plot_predictions(self):
        plt.figure(figsize=(20,4))
        plt.scatter(self.idata.constant_data.t, self.idata.constant_data.y, label='training data')
        plt.scatter(self.idata_preds.predictions_constant_data.t_fut, self.idata_preds.predictions.yhat_fut.mean(['chain','draw']), label='predicted values')
        plt.scatter(self.data_obj.t, self.data_obj.data, alpha=0.3, label='true values')
        plt.legend()
```

```python
class gen_synth_data_no_ar:

    def __init__(self, t_train=300,t_test=200 ):
        self.t_train = t_train
        self.t_test = t_test
        self.t_length = self.t_train+self.t_test
        np.random.seed(3)
        self.t = np.arange(0,self.t_length)

    def build_data_no_ar(self):

        self.linear_comp = self.sim_lin(slope=0.01, intercept=0)
        self.seasonal_comp = self.sim_cyclic(amplitude = 4, frequency=(1/50), phase=0)

        self.data = self.linear_comp+self.seasonal_comp + np.random.normal(0,1,self.t_length)

    
    def sim_cyclic(self, amplitude, frequency, phase):
        cyc = amplitude*np.sin(2*np.pi*frequency*(self.t + phase))
        return cyc
        
    def sim_lin(self, slope, intercept):
        lin_val = slope*self.t + intercept
        return lin_val
        
    def plot_custom_data(self):
        
        plt.figure(figsize=(20,5))
        plt.plot(self.t, self.data)
        plt.title('Data')
    
        plt.figure(figsize=(20,5))
        plt.plot(self.t, self.linear_comp)
        plt.title('linear')
    
        plt.figure(figsize=(20,5))
        plt.plot(self.t, self.seasonal_comp)
        plt.title('seasonal')
```
