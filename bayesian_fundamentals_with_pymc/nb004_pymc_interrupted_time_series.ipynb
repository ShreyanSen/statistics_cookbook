# Interrupted Time Series

in which we follow: https://www.pymc.io/projects/examples/en/latest/causal_inference/interrupted_time_series.html

this problem can be considered an extension of the kind of problem we solved in nb1. we have some data to which we applied an intervention at known time T, and we want to understand the effect of this intervention on our outcomes of interest. the basic approach we use is to model the data before time T, and then to use our model to generate an expectation for observations after time T. we'll track the difference between what was observed after time T based on what was expected based on our model for before time T, and attribute that difference to our intervention. In other words we compare the modeled counterfactual ("what would have happened without intervention") to the real data we have from our post-intervention.


first, let's generate our time series.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pymc as pm
import arviz as az
import xarray as xr
```

```python
RANDOM_SEED = 8227
rng = np.random.default_rng(RANDOM_SEED)
```

```python
treatment_time = "2017-01-31"
B0 = 0
B1 = 0.1
dates = pd.date_range(
    start=pd.to_datetime("2010-01-01"), end=pd.to_datetime("2020-01-01"), freq="M"
)
N = len(dates)
df = pd.DataFrame().assign(time=np.arange(N), date=dates).set_index("date", drop=True)

def causal_effect(df):
    return (df.index > treatment_time) * 5

df['y'] = B0 + B1*df.time + causal_effect(df)+ np.random.normal(0,1,N)

```

```python
ttime = df[df.index==treatment_time].time[0]
```

```python
# Split into pre and post intervention dataframes
pre = df[df.index < treatment_time]
post = df[df.index >= treatment_time]
```

```python
plt.figure(figsize=(12,5))
plt.plot(pre.time, pre.y, label='pre')
plt.plot(post.time, post.y, label='post')
plt.axvline(ttime, c="k", ls=":", label='treatment time')
plt.title('Outcomes Before and After Jump')
plt.xlabel('Months Since Start')
plt.ylabel('Outcome')
plt.legend()
```

```python
with pm.Model() as model:
    #time = pre.time # this would have normally been how I'd do it, but we want to update this time variable later for our counterfactual!
    time = pm.MutableData("time", pre.time)
    beta0 = pm.Normal("beta0", 0,1)
    beta1 = pm.Normal("beta1",0,1)
    mu = pm.Deterministic("mu",beta0 + beta1*time)
    sigma = pm.HalfNormal("sigma", 2)
    observation = pm.Normal("obs", mu=mu, sigma=sigma, observed = pre.y)
```

```python
with model:
    step = pm.Metropolis()
    trace = pm.sample(10000, tune=5000, step=step, return_inferencedata=True, random_seed=RANDOM_SEED)
```

```python
trace.posterior
```

## Check the fit

How are our param posteriors looking? All three params are looking pretty close to the truth

```python
plt.figure(figsize=(5,5))
plt.hist(trace.posterior.beta0.to_numpy().flatten())
plt.title('Beta0 Samples')

plt.figure(figsize=(5,5))
plt.hist(trace.posterior.beta1.to_numpy().flatten())
plt.title('Beta1 Samples')

plt.figure(figsize=(5,5))
plt.hist(trace.posterior.sigma.to_numpy().flatten())
plt.title('Sigma Samples')
```

## Prior Predictive Check

We want to make sure the real data are safely within the range of the kind of data generated from the prior, so that the model isn't overly constrained by the prior!

```python
with model:
    idata = pm.sample_prior_predictive(random_seed=RANDOM_SEED)
```

```python
plt.figure(figsize=(12,5))
az.plot_hdi(pre.index, idata.prior_predictive["obs"], hdi_prob=0.4, smooth=False)
az.plot_hdi(pre.index, idata.prior_predictive["obs"], hdi_prob=0.95, smooth=False)
plt.plot(pre.index, pre["y"], label="observed")
plt.legend()
plt.title('Prior Predictive Observations vs. Observed Data')
```

## Posterior Predictive Check

We do the same thing with the posterior predictive values to see whether our posterior does a good job capturing the original data. It does!

```python
with model:
    idata.extend(pm.sample_posterior_predictive(trace, random_seed=RANDOM_SEED))
```

```python
plt.figure(figsize=(12,5))
az.plot_hdi(pre.index, idata.posterior_predictive["obs"], hdi_prob=0.5, smooth=False)
az.plot_hdi(pre.index, idata.posterior_predictive["obs"], hdi_prob=0.95, smooth=False)
plt.plot(pre.index, pre["y"], label="observed")
plt.legend()
plt.title('Posterior Predictive Observations vs. Observed Data')
```

Well, it looks like the tutorial code works just fine! I suspect my issues have to do with not defining the dimensionality of my model components. In NB3 this was also a major theme! A great example of how syntax and theory can come together in weird ways that deeply matter.

```python
# just drop defining dimensions and see what happens

with pm.Model() as model3:
    # observed predictors and outcome
    time = pm.MutableData("time", pre["time"].to_numpy())
    # priors
    beta0 = pm.Normal("beta0", 0, 1)
    beta1 = pm.Normal("beta1", 0, 0.2)
    # the actual linear model
    mu = pm.Deterministic("mu", beta0 + (beta1 * time))
    sigma = pm.HalfNormal("sigma", 2)
    # likelihood
    pm.Normal("obs", mu=mu, sigma=sigma, observed=pre["y"].to_numpy())

with model3:
    idata3 = pm.sample_prior_predictive(random_seed=RANDOM_SEED)

with model3:
    idata3.extend(pm.sample(random_seed=RANDOM_SEED))

with model3:
    idata3.extend(pm.sample_posterior_predictive(idata3, random_seed=RANDOM_SEED))

with model3:
    pm.set_data(
        {
            "time": post["time"].to_numpy(),
        }
    )
    counterfactual = pm.sample_posterior_predictive(
        idata3, var_names=["obs"], random_seed=RANDOM_SEED
    )
```

### Result of Syntax Change

Yup! We hit the same error just by removing the dimension component. Let's look at the idata vs idata3 object below (one where we defined dimensions, one where we didn't). We see that the posterior predictive of idata has the same third dimension name as the posterior does, whereas this is not the case for idata3. Does this mean something significant? It must. Or at least, the dimensions for time must be set to match the dimensions for pm.Normal. 


I experimented by removing dim="obs_id" or adding it back in for a subset of the variables we have it in in the working code instance. Removing dimensionality on mu and sigma doesn't break the code in the same way. If I keep the dimensionality on pm.Normal but remove the dimensionality on my pm.MutableData time value, the code still breaks, complaining that "obs_id" is not defined and is an unknown label! So, it's important to connect the mutable data object's dimensionality to the observation dimensionality for predicting forward in time! And somehow you're defining "obs_id" as a valid label in that MutabaleData object. Rather confusing!


Some words on syntax. It's a good thing to become familiar with your API of choice, but it's more important to understand the modeling fundamentals. If we couldn't figure out the API, could we still have made this prediction? Yes! We could have done it manually. How? We could have drawn samples for our parameters beta1 and beta0, then multiplied sets of these params with our new time values, then computed mu and sigma and drawn an observation. Looping over this procedure we could have manually generated sets of posterior predictive observations for t > T.

## Complete the Tutorial: Causal Impact

Our basic approach will be to get the residuals of the real vals minus the predicted vals (from the pre intervention model) and look at their effect and cumulative effect

```python
observed_vals = post.y.to_numpy()
counterfactual_vals = counterfactual.posterior_predictive["obs"].to_numpy()
```

```python
impact_vals = np.subtract(observed_vals, counterfactual_vals)
np.shape(impact_vals)
```

```python
cum_impact = np.cumsum(impact_vals, axis=2)
```

```python
plt.figure(figsize=(12,5))
az.plot_hdi(post.index, impact_vals, hdi_prob=0.5, smooth=False)
az.plot_hdi(post.index, impact_vals, hdi_prob=0.95, smooth=False)
plt.ylabel('Attributable Difference')
plt.legend()
plt.title('Real Minus Counterfactual Residuals')
```

```python
plt.figure(figsize=(12,5))
az.plot_hdi(post.index, cum_impact, hdi_prob=0.5, smooth=False)
az.plot_hdi(post.index, cum_impact, hdi_prob=0.95, smooth=False)
plt.ylabel('Cumulative Attributable Difference')
plt.legend()
plt.title('Cumulative Real Minus Counterfactual Residuals')
```

## Comment on "Causal"

Is this difference really causally attributable. No, not strictly. Why? Anything else that changed on our intervention date could have driven this change in outcome. For example, had our company haplessly rolled out their change on 9/11/2001, this kind of analysis would have been of course deeply flawed. However, if we believe that nothing else about our underlying process changed before and after our intervention date, then it's reasonable to attribute the observed difference to our intervention, even if we can't make a strict causal claim about it.
